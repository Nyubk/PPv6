```python
#Modulos
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
lluvia0020_ = pd.read_csv('https://github.com/Nyubk/PPv6/raw/refs/heads/main/CNPV2000_2020/precipitacion00-20.csv')
```

```python
#Copia de seguridad
lluvia0020 = lluvia0020_
```

```python
# Ajustar el formato del dataset para el mapa de calor
lluvia0020_cleaned = lluvia0020.set_index(lluvia0020.columns[0])  # Usar la primera columna como índice
```

```python
# Generar el mapa de calor
plt.figure(figsize=(12, 6))
sns.heatmap(lluvia0020_cleaned.astype(float), annot=True, cmap='YlGnBu', cbar=True)
plt.title("Mapa de Calor de Precipitación (2000-2020)")
plt.xlabel("Meses")
plt.ylabel("Región/Año")
plt.tight_layout()
plt.show()
```
